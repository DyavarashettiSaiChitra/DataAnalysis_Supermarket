```python
#changed Order date('Obj') date type to datetime datatype
import pandas as pd

sales['Order Date'] = pd.to_datetime(sales['Order Date'], format='mixed')   
```

```python
#Created year,month,day columns to analyse sales as per Year,month
sales['Year'] = sales['Order Date'].dt.year
sales['Month'] = sales['Order Date'].dt.month
sales['Day'] = sales['Order Date'].dt.day
```

```python
#Data visualisation
#Sales by category
import matplotlib.pyplot as plt
data=sales.groupby(['Category','Year'])['Sales'].sum().unstack()
data.plot(kind='bar',figsize=(10,6),color=['#9C6DB0','#5AC494','#A77680','#DBE637'])
plt.xlabel('category',fontsize=15,color='green')
plt.ylabel('sales',fontsize=15,color='green')
plt.grid(color='green',linestyle=':',linewidth=1)
plt.tight_layout()
plt.xticks(rotation=0)
plt.legend(loc='upper left',fontsize=5)
plt.show()
```

```python
pivot = pd.pivot_table(sales,           #dataframe/dataset
                       values='Sales',  #col to calculate
                       index='Category', #row_category
                       columns='Year',    #col_category
                       aggfunc='sum')  #function-mean,sum,count
print(pivot)
```

```python
pivot.plot(kind='bar', figsize=(8,5))
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Year-wise Sales Comparison by Category')
plt.xticks(rotation=90)
plt.legend(loc='upper left',fontsize=8)
plt.show()
```

```python
#TREND ANALYSIS
pivot.plot(kind='line', marker='o', figsize=(8,5),color=('#9C6DB0','#5AC494','#A77680','#DBE637'))
plt.title('Sales Trend by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.grid(linestyle=':')
plt.xticks(rotation=90)
plt.show()
```

```python
#TOTAL SALES BY YEAR(stacked bar chart)
pivot.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title('Stacked Sales by Category and Year')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#Category wise sales(pie chart)
category_sales = sales.groupby('Category')['Sales'].sum()
plt.pie(category_sales,labels=category_sales.index,autopct='%1.1f%%')
plt.title('Sales Contribution by Category')
plt.show()
```

```python
#complete dashboard --->data cleaning,groupby analysis,aggregation,data visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
#Sales by Category 
plt.subplot(2,2,1) #partition of whole chart
category_sales = sales.groupby('Category')['Sales'].sum()
plt.bar(category_sales.index, category_sales.values, color='purple')
plt.title('Sales by Category')
plt.xticks(rotation=45)

#Profit by Region
plt.subplot(2,2,2)
region_profit = sales.groupby('Region')['Profit'].sum()
plt.bar(region_profit.index, region_profit.values, color='green')
plt.title('Profit by Region')

#Year-wise Sales Trend
plt.subplot(2,2,3)
year_sales = sales.groupby('Year')['Sales'].sum()
plt.plot(year_sales.index, year_sales.values, marker='o')
plt.title('Year-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')

#Sales Contribution Pie Chart
plt.subplot(2,2,4)
plt.pie(category_sales,labels=category_sales.index,autopct='%1.1f%%')
plt.title('Sales Distribution by Category')
plt.tight_layout()
plt.show()
```
